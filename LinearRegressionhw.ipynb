```python
from  sklearn.linear_model import LinearRegression
#sample data
year=[[0],[1],[2],[3],[4]]
salary = [4000,8000,10000,12000,20000]

#create model and train
model = LinearRegression()
model.fit(year,salary)

#predict salary for 5 year
predict_salary = model.predict([[5]])
print("predicted salary for 5 year :" ,predict_salary[0],"lacks")
```

```python
import matplotlib.pyplot as plt
plt.scatter(year,salary,color ='blue',label='actual data')
plt.plot(year,model.predict(year),color ='red', label='regression ')
plt.scatter([[5]], predict_salary , color = 'green', label =' predicted(5 year)')
plt.xlabel('year')
plt.ylabel('salary')
plt.legend()
plt.grid(True)
plt.show()
```
