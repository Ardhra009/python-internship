```python
from  sklearn.linear_model import LinearRegression

area=[[500],[1000],[1500],[2000],[2500]]#sqft
price=[40,80,120,160,200]#in lakhs
model=LinearRegression()
model.fit(area,price)
predicted_price=model.predict([[3000]])
print("predicted price for 3000 sqft house:",predicted_price[0],"lakhs")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(area,price,color='blue',label='actual data')
plt.plot(area,model.predict(area),color='red',label='Regression Line')
plt.scatter([[3000]],predicted_price,color='green',label='predicted (3000 sqft)')
plt.xlabel('Area (sqft)')
plt.ylabel('Price (lakhs)')
plt.legend()
plt.grid(True)
plt.show()

```
