```python
import pandas as pd
data={"DEPARTMENT":["accountant","mechanical","software"],"EMPLOYEE":["lina","kin","hyla"],"SALARY":[67000,85000,25000]}
df=pd.DataFrame(data)
df
```

```python
avg_salary=df.groupby("DEPARTMENT")["SALARY"].mean()
print(avg_salary)
```

```python
emp_count=df.groupby("DEPARTMENT")["EMPLOYEE"].count()
print(emp_count)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([10,20,25,30])
plt.xlabel('day')
plt.ylabel('users')
plt.title('daily users')
plt.plot(x,y)
plt.show()
```

```python
product=np.array(['pen','book','eraser'])
sales=np.array([100,150,80])
plt.bar(product,sales)
plt.xlabel('products')
plt.ylabel('sales')
plt.show
```

```python
height=np.array([150,160,170,180])
weight=np.array([50,60,65,75])
plt.scatter(height,weight)
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
shares=np.array([30,25,20,25])
smartphone=['apple','samsung','xiomi','others']
mycolors=['black','blue','pink','#4CAF50']
plt.pie(shares,labels=smartphone,colors=mycolors)
plt.show()
```
